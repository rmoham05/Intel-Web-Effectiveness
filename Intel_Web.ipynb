```python
# Required packages for working with dataframe and visulizing it

import numpy as np 

import matplotlib.pyplot as plt

import pandas as pd 

import seaborn as sns
```

```python
df.info()
```

<h3>As it can be seen in the table above our dataset consists of 116208 rows (samples) and 24 columns (features) </h3>

<h3>For the second step we should get rid of the columns in which more than %50 of data is Null and fill the rest of Null values with Unknown</h3>

```python
df.drop(["Improve site to better meet needs","Purchase decision process","Planned spending","Intel.com makes it easy to determine where to purchase","Adoption","Frequency of visit"], axis=1, inplace=True)
```

```python
df.info()
```

```python
# for each column, get value counts in decreasing order and take the index (value) of most common class
df_cleaned = df.apply(lambda x: x.fillna("Unknown"))
df_cleaned
```

```python
df_cleaned.info()
```

<h3>One of the primary reason to show the effectiveness of a website can be found by checking if the customers are satisfied or not </h3>

<h3>So let's check the proportion of customers based on their satisfaction level</h3>

```python
satisfaction=df_cleaned["Satisfaction"].value_counts()
satisfaction
```

<h4>As it can be seen more than %68 of customers are satisfied and just %12 are dissatisfied</h4>

```python
pie, ax = plt.subplots(figsize=[10,6])
labels = satisfaction.keys()
plt.pie(x=satisfaction, autopct="%.1f%%", explode=[0.05]*3, labels=labels, pctdistance=0.5)
plt.title("Satisfaction", fontsize=14);
```
